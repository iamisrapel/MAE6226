```python
import numpy
from matplotlib import pyplot
import math
%matplotlib inline
```

```python
x = numpy.linspace(-20,20,101)
y = numpy.linspace(-10,10,101)
X,Y = numpy.meshgrid(x,y)
```

```python
gamma = 5.
Ngedian = 20
x_vortex = numpy.linspace(-20,20,Ngedian)-0.1
y_vortex = numpy.ones(Ngedian)
```

```python
u = numpy.zeros([101,101])
v = numpy.zeros_like(u)
psi = numpy.zeros_like(u)
```

```python

for i in range(Ngedian):

    u_linshi = +gamma/(2*math.pi) *(Y-y_vortex[i])/((X-x_vortex[i])**2 + (Y - y_vortex[i])**2)
    v_linshi = -gamma/(2*math.pi) *(X-x_vortex[i])/((X-x_vortex[i])**2 + (Y - y_vortex[i])**2)
    psi_linshi = gamma/(4*math.pi) * numpy.log((X-x_vortex[i])**2 + (Y - y_vortex[i])**2)

    u = u+u_linshi
    v = v+v_linshi
    psi = psi +psi_linshi
```

```python
pyplot.figure(figsize=(10,4))
pyplot.streamplot(X,Y,u,v,density=5, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.xlim(-5,5)
pyplot.ylim(-1,3)
pyplot.scatter(x_vortex, y_vortex)
```

```python
uu = gamma/(2*1) *numpy.sinh(2*math.pi*Y/1)/(numpy.cosh(2*math.pi*Y/1)-numpy.cos(2*math.pi*X/1))
vv = -gamma/(2*1) *numpy.sin(2*math.pi*X/1)/(numpy.cosh(2*math.pi*Y/1)-numpy.cos(2*math.pi*X/1))
```

```python
pyplot.figure(figsize=(10, 10))
pyplot.streamplot(X,Y,uu,vv,density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.xlim(-10,10)
pyplot.ylim(-10,10)
```

```python
pyplot.figure(figsize=(10, 10))
pyplot.streamplot(X,Y,uu,vv,density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.xlim(-10,10)
pyplot.ylim(-2,2)
```
