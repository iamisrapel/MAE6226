```python
import numpy
import math
from matplotlib import pyplot
```

```python
N = 50
x = numpy.linspace(-2., 2., 51)
y = numpy.linspace(-1., 1., 51)
X, Y = numpy.meshgrid(x, y)
```

```python
%matplotlib inline
```

```python
pyplot.figure(figsize=(10,5))
pyplot.xlim(-2, 2)
pyplot.ylim(-1, 1)



pyplot.scatter(X, Y, s=2, color = '#CD0001', marker = 'o',linewidth=0 )
```

```python
strength_source = 5.
x_source, y_source = -1, 0.
```

```python
u_source = strength_source/(2*math.pi)*(X - x_source)/((X-x_source)**2+(Y-y_source)**2)
v_source = strength_source/(2*math.pi)*(Y - y_source)/((X-x_source)**2+(Y-y_source)**2)
```

```python
pyplot.figure(figsize = (10,5))
pyplot.xlim(-2,2)
pyplot.ylim(-1,1)
pyplot.streamplot(X,Y,u_source, v_source, density = 3, linewidth = 1, arrowsize = 1, arrowstyle='->')
pyplot.scatter(x_source, y_source,color='#CD2305',s=80, marker='o',linewidth=0 )
```
