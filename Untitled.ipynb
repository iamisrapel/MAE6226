```python
import numpy
```

```python
x=numpy.loadtxt('./NACA0012_x.txt')
y=numpy.loadtxt('./NACA0012_y.txt')
x=x-16
```

```python
from matplotlib import pyplot
%matplotlib inline
```

```python
pyplot.plot(x,y)
```

```python
xi=x+1j*y
```

```python
z=0.5*(xi-(xi**2 - 4)**0.5)
```

```python
x_z=z.real
y_z=z.imag
```

```python
pyplot.plot(x_z,y_z)
```
